```python
# üìö Cell 2: Import Required Libraries
from transformers import pipeline
from newspaper import Article, ArticleException # Import ArticleException
import pdfplumber
from google.colab import files

# ... (rest of your code) ...

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_text_in_chunks(text, chunk_size=700, min_len=100, max_len=250):
    # Break text into chunks
    import textwrap
    chunks = textwrap.wrap(text, chunk_size)

    full_summary = ""

    for i, chunk in enumerate(chunks):
        print(f"‚è≥ Summarizing chunk {i+1}/{len(chunks)}...")
        try:
            summary = summarizer(chunk, min_length=min_len, max_length=max_len, do_sample=False)
            full_summary += summary[0]['summary_text'] + " "
        except Exception as e:
            print(f"‚ùå Error summarizing chunk {i+1}: {e}")

    return full_summary.strip()

def summarize_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text
    return summarize_text_in_chunks(text)


import time

def summarize_url(url, retries=3, delay=5):
    article = Article(url)
    for i in range(retries):
        try:
            article.download()
            article.parse()
            return summarize_text_in_chunks(article.text)
        except ArticleException as e:
            if i < retries - 1:
                print(f"‚ùå Download failed, retrying in {delay} seconds... (Attempt {i+1}/{retries})")
                time.sleep(delay)  # Wait before retrying
            else:
                print(f"‚ùå Download failed after {retries} attempts: {e}")
                # Instead of raising the exception, return an error message
                return "‚ùå Could not download or summarize the article. Please check the URL and try again."

print("Select input type:")
print("1 - Plain Text")
print("2 - PDF File")
print("3 - URL")

choice = input("Enter your choice (1/2/3): ")

if choice == '1':
    # Plain Text
    text = input("Paste your long text below:\n")
    summary = summarize_text_in_chunks(text) # Call the correct function for summarizing plain text
    print("\nüìå Summary:\n", summary)

elif choice == '2':
    # PDF File Upload
    print("üì§ Upload your PDF file")
    uploaded = files.upload()
    file_name = list(uploaded.keys())[0]
    summary = summarize_pdf(file_name)
    print("\nüìå Summary:\n", summary)

elif choice == '3':
    # Article URL
    url = input("Paste the article URL:\n")
    summary = summarize_url(url)
    print("\nüìå Summary:\n", summary)

else:
    print("‚ùå Invalid choice. Please enter 1, 2, or 3.")
```
